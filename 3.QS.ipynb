#### 3. Write a program to check whether the given number is in between 1 and 100 ?

```python
a=int(input())
if a in range(1,100):
    print('The number is in between 1 to 100')
else:
    print('The number is not in between 1 to 100')
```

#### 5. Write a program to take a single digit number from the key board and print is value in English word ?

```python
a=int(input())
b=1
while a+1>b:
    print('*'*b)
    b=b+1
```

#### 6. (Alternate Way) - While – Loop : (Write a program to display *’s in right Angled Triangled Form ?

```python
a=int(input())
b=1
while a+1>b:
    c=a-b
    print(' '*c,'*'*b)
    b=b+1
```

#### 7. Write a program to display *’s in pyramid style (Also known as Quivalent Triangle) ?


```python
a=int(input())
b=1
while a+1>b:
    c=a-b
    if b==1:
        print(' '*c,'*'*b,)
    if b>1:
        print(' '*c,'*'*b+'*'*(b-1))
        
    b=b+1

    
```

#### 12. Write a program to access each character of string in Forward and Backward Direction by using while loop ?

```python
a=input()
b=0
while b<len(a):
    print(a[b],' ',a[-(b+1)])
    #print(a[-(b+1)])
    b=b+1
```
